We plan on considering using the following 3 algorithms for our project:


1.   **k Nearest Neighbors**
    <br>k Nearest Neighbors operates under the assumption that similar points of
    data exist in close proximity, essentially, birds of a feather flock together. kNN looks at all the specific points closest to the supplied point to predict, and counts the "votes" for the classification that appears the most. Since our data is not sparse, it is reasonable to assume that it would work well with our dataset since it would help us classify a supplied point and there is a lot of data to work with, hence ensuring an accurate classification with the calculated predictors and patterns within the data. </br>

2.   **Decision Tree**
    <br>Decision trees assume that the entire training set is the "root" at the start of the tree/algorithm, and proceeds to categorize data in the set based on "tests" of whether data has a specific attribute or not. With 10 feature variables, we would be able to classify each data point easily because of the in-depth features we are using for the training set. Any nodes with similar attributes would be placed together. The algorithm would learn based on the decisions/categories the algorithm has made and subsequently be able to predict how well a new anime show would fare in terms of user review.</br>

3.   **Naive Bayes**
    <br>Naive Bayes operates under the assumption that all features are equally prioritized in the classification system and all features are statistically independent. Subsequently, the algorithm uses probabilistic classifiers to partition the data; thus, similar points end up in the same classifier. Based on the probability that a supplied data point to predict falls into a certain category, we can predict which category that a new data point would fall into. Since we have many data points that share similar values in terms of one (or more) of our feature variables, using Naive Bayes would help us split the data up into groups of data points that are similar to one another.</br>

### 3.1. Data Wrangling


```python
#read data into df
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/caitchao/ds3000-project-data-anime/main/anime_data.csv', index_col=0)
df
```



> These functions will be used to map onto certain columns of the dataframe for easier manipulation of the features.



```python
from datetime import datetime, date
#given a date of birth in the MM/DD/YYYY format the function calculates
#the age based on the date
def date_to_age(born):
    try:
        born = datetime.strptime(born, "%m/%d/%Y").date()
    except:
        return float("NaN")
    today = date.today()
    return today.year - born.year - ((today.month, today.day) < (born.month, born.day))

import string
#given a string representation of a list of genres, the function returns
#the first genre (main genre) of the list as a string
def get_dom_genre(s):
    for char in string.punctuation:
        if char != "-":
            s = s.replace(char, ' ')
    l = s.split()
    if len(l) > 0:
        return l[0]
    else:
        return float("NaN")
```

Functions get mapped onto columns, and column names are changed appropriately to prepare for feature preprocessing. 

```python
#map column for birthday to age
df["birthday"]=df["birthday"].map(date_to_age)
df = df.rename(columns={"birthday": "age"})

#map column for genre list to main genre
df["genre"]=df["genre"].map(get_dom_genre)

df = df.dropna()
df
```

Now we one-hot encode the gender and genre, both which are both categorical variables, to represent them in terms of 0s and 1s. We do this to make the cateogircal variables easier to work with, since numerical values are easier to work with. 

```python
#categorical columns to one-hot encode
features_a = df.loc[:, ['gender', 'genre']]

from sklearn.preprocessing import OneHotEncoder

#create the encoder
encoder = OneHotEncoder(sparse = False)

#encode the given columns
encoded_df = encoder.fit_transform(features_a)

#create dataframe with encoded columns
features_a = pd.DataFrame(encoded_df, columns = encoder.get_feature_names_out(["gender", "genre"]))

features_a
```

Next, we use the standard scaler on three of the variables-- episodes, popularity and rank. This is to standardize the values in these columns so that the mean is 0 and standard deviation is 1. We do this so that the machine learning model does not prioritize one variable over another. 

The profile, favorite_anime, uid, and title feature variables are not going to be used for training because they are arbitrary. The scores target variable column is also not going to be used because it provides multiple categories for target variables, which all can be gauged through the overall_score column.

```python
# columns that we want to apply standard scaler on
features_b = df.loc[:, ['episodes', 'popularity', 'rank']]

from sklearn.preprocessing import StandardScaler

#create the scaler
scaler = StandardScaler()

#fit the scaler to the training data(certain features only)
scaler.fit(features_b) 

#transform features based on the (same) scaler
features_b = scaler.transform(features_b) 
features_b = pd.DataFrame(features_b, columns = scaler.get_feature_names_out(['episodes', 'popularity', 'rank']))

features_b
```

We now put together the pre-processed variables and original variables together with the target feature. 

```python
#putting data samples together after preprocessing
age_df = df.loc[:, ['age']]
scores_df = df.loc[:, ['overall_score']]
df_final = pd.concat([age_df, features_a, features_b, scores_df], axis=1)

df_final
```

We then perform data cleaning on this final dataframe by removing any null values and rows with irregular data (for example if age of a user is less than or equal to 0).

```python
#data cleaning
cleaned_df = df_final.dropna()
cleaned_df = cleaned_df.reset_index()
cleaned_df = cleaned_df.drop(columns=["index"])
cleaned_df = cleaned_df[cleaned_df["age"]>0]

cleaned_df
```

After splitting our dataframe into the feature variables and the target variable, we use train/test split to randomly split our dataframe into the training and testing sets. 

```python
#split columns into features and target
features = cleaned_df.drop("overall_score", axis=1)
target = cleaned_df["overall_score"]
```

```python
from sklearn.model_selection import train_test_split

#split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, target, random_state=3000)
```

### 3.3. Model Training


```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB

from sklearn.metrics import accuracy_score, f1_score

# dictionary contains three classifiers
classifiers = {"k-Nearest Neighbor": KNeighborsClassifier(), 
              "Decision Tree": DecisionTreeClassifier(), 
              "Gaussian Naive Bayes": GaussianNB()}
```

We chose to use accuracy and f1 score metrics to evaluate our algorithm performance. Accuracy can be used to determine how well the algorithm classifies, and the f1 score can be used to gague the classifiers' exactness and completeness. With both of these metrics, we can get a sense of the model's effectiveness by evaluating the scores based on True/False Positives

```python
# iterate through dictionary of classifiers
for name, clf in classifiers.items():
    
    #create the model by fitting the training data
    clf.fit(X=X_train, y=y_train)
    
    # making predictions
    pred = clf.predict(X_test)
    
    #prediction accuracy
    accuracy = accuracy_score(y_test, pred)
    fone_score = f1_score(y_test, pred, average='weighted')

    print(f"{name}:")
    print("\tAccuracy score on the test data: ", accuracy)
    print("\tF1 score on the test data: ", fone_score,"\n")
```

Next, we will try to perform feature selection using Recursive Feature Selection to select a subset of our original features and train the models with the selected features, so we can see if we can achieve the same using a fewer number of features. However, since we used one-hot encoding for two of our categorical variables, we will exclude them from this feature selection test, and try to feature select three variables from the age, episodes, rank, and popularity variables. 

```python
X_train_toselect = X_train[["age", "episodes", "rank", "popularity"]]
X_test_toselect = X_test[["age", "episodes", "rank", "popularity"]]
```

```python
from sklearn.feature_selection import RFE
from sklearn.tree import DecisionTreeRegressor

def RFE_feature_selection():

    # Create a RFE object
    select = RFE(DecisionTreeRegressor(random_state=3000), n_features_to_select = 3)

    # Fit the data to the RFE
    select.fit(X_train_toselect, y_train)

    # transform the scaled data with the selected features
    X_train_selected = select.transform(X_train_toselect)
    X_test_selected = select.transform(X_test_toselect)

    column_names = X_train_toselect.columns.values

    # output the selected features where the support value is True 
    print("Selected features after RFE:")
    for x in range(len(select.support_)):
      if select.support_[x] == True:
        print(column_names[x])

    return X_train_selected, X_test_selected 
```

```python
X_train_selected, X_test_selected = RFE_feature_selection()
```

```python
# iterate through dictionary of classifiers
for name, clf in classifiers.items():
    
    #create the model by fitting the training  data
    clf.fit(X=X_train_selected, y=y_train)
    
    # making predictions
    pred = clf.predict(X_test_selected)
    
    #prediction accuracy
    accuracy = accuracy_score(y_test, pred)
    fone_score = f1_score(y_test, pred, average='weighted')

    print(f"{name}:")
    print("\tAccuracy score on the test data: ", accuracy)
    print("\tF1 score on the test data: ", fone_score,"\n")
```

From these results, we can see that using three of the features selcted through Recursive Feature Selection analysis produces similar or even enhanced results to the original set of features used, with higher accuracy and f1 scores.

### 3.4. Model Optimization


We would now like to tune our models using hyperparameters that are appropriate for each model. From the above data, we decided to proceed with Decision Tree model and the Gaussian Naive Bayes model, which had higher accuracy rates than k-Nearest Neighbors. We use hyperparameter tuning in an attempt to minimize overfitting, or matching too closely to the training set. We try to use hyperparameter tuning by fitting to our original feature variables and also to our selected features, in an attempt to optimize performance. We encountered a UserWarning when using cross validation in our GridSearchCV, but this did not interfere with obtaining results.

```python
# import GridSearchCV and fit GridSearchCV
from sklearn.model_selection import GridSearchCV

# possible values for each hyperparameter
param_grid = {"max_depth":[10, 100, 1000, 10000, 10000],
             "max_leaf_nodes": [10, 100, 1000, 10000, 100000]}

# apply Grid Search with decision tree classifier and evaluate performance of parameters with 5-fold evaluation
grid_search = GridSearchCV(DecisionTreeClassifier(), param_grid, cv=5)

#fit the grid search object on the training data 
grid_search.fit(X=X_train, y=y_train)

report = ''
# result of grid search
report += "Decision Tree cross-validation\n"
report += f"Best Cross-validation score:  {grid_search.best_score_}\n" 
report += f"Best parameters:  {grid_search.best_params_}\n"

# the performance of the best found parameters on the training 
report += f"Training set score with best parameters:  {grid_search.score(X_train, y_train)}\n"
    
print (report)
```

## 4. DISCUSSION



  For our data, we used 3 algorithms in order to compare the results of each one: k Nearest Neighbors, Decision Trees, and Naive Bayes. We redid the model training and testing and our results yielded similar numbers to our results sans feature selection.

  Out of the 3 algorithms we used, we proceeded to perform hyperparameter tuning on decision trees and Naive Bayes to avoid overfitting our data.
