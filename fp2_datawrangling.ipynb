```python
import pandas as pd
df_anime = pd.read_csv('res/animes.csv')
df_anime = df_anime.drop_duplicates()
df_anime
```

```python
df_profile = pd.read_csv('res/profiles.csv')
df_profile = df_profile.drop_duplicates()
df_profile
```

```python
df_review = pd.read_csv('res/reviews.csv')
df_review = df_review.drop_duplicates()
df_review
```

```python
profile_review = pd.merge(df_profile, df_review, on="profile")
cols = ["profile", "gender", "birthday", "favorites_anime", "anime_uid", "score", "scores"]
profile_review = profile_review[cols]
profile_review = profile_review.rename(columns={"anime_uid": "uid"})
```

```python
df_all = pd.merge(profile_review, df_anime, on="uid")
cols = ["profile", "birthday", "gender", "favorites_anime", "uid", "title", "genre", "episodes", "popularity", "ranked", "scores", "score_x" ]
df_all = df_all[cols]
df_all = df_all.drop_duplicates()
df_all = df_all.rename(columns={"ranked": "rank", "score_x": "overall_score"})
df_all
```

```python
import string
def remove_chars(s):
    for char in string.punctuation:
        if char != "-":
            s = s.replace(char, ' ')
    return s

def uidstring_to_list(s):
    l = s.split("'")
    l = [int(x) for x in l if x.isnumeric()]
    return l

def anime_list_to_genres(uids):
    genres = []
    uid_list = uidstring_to_list(uids)
    if len(uid_list) != 0:
        for uid in uid_list:
            genreString = df_anime.loc[df_anime['uid'] == int(uid),'genre']
            genreString = genreString.iloc[0]
            current_genres = genreString.split("'")
            current_genres = [x for x in current_genres if x.upper().isupper()]
            for g in current_genres:
                if g not in genres:
                    genres.append(g)
    return genres
```

```python
df_all["favorites_anime"]=df_all["favorites_anime"].map(anime_list_to_genres)
df_all
```

```python
cleaned = df_all.dropna()
cleaned = cleaned.reset_index()
cleaned = cleaned.drop(columns=["index"])
cleaned
```

```python
cleaned.to_csv("res/fp2.csv")
```
